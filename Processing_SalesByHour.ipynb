```python
import pandas as pd
import numpy as np

df = pd.read_csv("SalesByHour.csv",parse_dates=['Date'], low_memory=False)
#df['Department (no hierarchy)'] = df['Department (no hierarchy)'].str.split(' - ', n=1).str[1]
df = df.rename(columns={'[ES] Category': 'Category'})
```

```python

#df.dtypes
# Convert 'Sum of Quantity' to float64

#df = df[df['Date'] != 'Overall Total']
#df['Sum of Quantity'] = pd.to_numeric(df['Sum of Quantity'], errors='coerce')

# Ensure 'Date' is of datetime type (this should be handled by parse_dates, but good to confirm)
#df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df[df['TimeStamp'].notna()]
# Remove columns 'UPC Code' and 'Display Name'
df = df.drop(columns=['UPC Code', 'Display Name','[ES] Sub-Department','[ES] Sub-Category','[ES] Type of Brand','Retail Price','Purchase Price','Sum of Quantity','[ES] Segment','[ES] Brand','[ES] Unit Cost','Preferred Vendor','Not For Ordering'])
df
```

```python
#df = df.rename(columns={'Location (no hierarchy)': 'Location'})
df['Location (no hierarchy)'] = df['Location (no hierarchy)'].replace("1001-Pusateri's Ltd. (Ave)",'Alameda')
df['Location (no hierarchy)'] = df['Location (no hierarchy)'].replace("1004-Pusateri's Bayview Ltd.",'Bayside')
df['Location (no hierarchy)'] = df['Location (no hierarchy)'].replace("1002-Pusateri's Yorkville Ltd.",'Yellowknife')
df['Location (no hierarchy)'].unique()
```

```python
df
df.to_csv('SalesByHour2.csv', index=False)
```
